```
def cleaning(data):
  x=[]
  x.append(data)
  for i in range(len(x)):
    x[i] = re.sub(r'\[[0-9]*\]',' ',x[i])
    x[i] = re.sub(r'\[[a-z]*\w+]',' ',x[i])
    regex = re.compile('[a-zA-Z]')
    #First parameter is the replacement, second parameter is your input string \uf06e \uf06c \u200c্যা 	new_string = re.sub("[^a-zA-Z]","", old_string)
    x[i] = regex.sub('', x[i])
    x[i] = re.sub("\uf06e", "", x[i])
    x[i] = re.sub("\uf06c", "", x[i])
    x[i] = re.sub("\u200c্যা", "", x[i])
    x[i] = re.sub("\u200c", "", x[i])
    x[i] = re.sub(r'[@#\$%&\*\(\)\<\>\?\'\‘\’\—\।\!\–\/\".,:;\]\[-]',' ',x[i])
    x[i] = re.sub(r'\d',' ',x[i])
    x[i] = re.sub(r'\s+',' ',x[i])
  return x[0]
```
